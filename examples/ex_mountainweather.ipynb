# Example on how to extract, plot and store a timeseries using the MountainWeather class

```python
__author__ = 'kmunve'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Make sure varsomdata is in your Pythonpath
import sys, os
sys.path.append(os.path.dirname(os.path.abspath(os.getcwd())))

from varsomdata import getforecastapi as gf
```

```python
date_range = pd.date_range(start='1/11/2018', end='28/1/2019')
#region_id = 3011  # Tromsø
region_id = 3010  # Lyngen
```

```python
df = pd.DataFrame(columns=['date_valid', 'temperature_max', 'temperature_min', 'precipitation_exposed', 'precipitation_avg', 'wind_spd', 'wind_dir'])
df
```

```python
for d in date_range:
    mw = gf.MountainWeather()
    mw.from_api_region(region_id, f'{d.year}-{d.month}-{d.day}')
    df = df.append({'date_valid': mw.date_valid,
                    'temperature_max': mw.temperature_max,
                    'temperature_min': mw.temperature_min,
                    'precipitation_exposed': mw.precip_most_exposed,
                    'precipitation_avg': mw.precip_region,
                    'wind_spd': mw.wind_speed,
                    'wind_dir': mw.wind_direction}, ignore_index=True)
        
df
```

```python
df.to_excel('mountainweather.xlsx', sheet_name='Lyngen')
```
